```python
from sklearn.neighbors import NearestNeighbors
```

```python
model_knn = NearestNeighbors(metric='cosine', algorithm='brute')
model_knn.fit(A.T)
```

```python
def get_item_recommendation(item_id, n_recommendations):
    # below method returns 2 arrays: 1. distances of nearest neighbours and 2.indices of nearest neighbours
    item_id= str(item_id)
    if item_id not in A.T.index:
        raise ValueError(f"Item ID {item_id} not found in the index.")
    distances, indices= model_knn.kneighbors(A.T.loc[item_id].values.reshape(1,-1), n_neighbors=n_recommendations+1)
    recommendations = [A.columns[indices.flatten()[i]] for i in range(1, len(distances.flatten()))]
    return recommendations
    
```
